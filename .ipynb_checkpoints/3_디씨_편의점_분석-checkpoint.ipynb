# 데이터 전처리

## 1. 정규 표현식

```python
# 새로운 컬럼 만들기
dc['글'] = dc['게시물제목']+ " " + dc['내용']

# 문자열str로 지정
dc['글'] = dc['글'].astype('str')

# 확인
dc.head()
```

```python
# 그 중에서 한글만 추출한, 새로운 컬럼 만들기 

# 한글만 추출해내기 = 정규표현식 = re 라이브러리 설치 
import re

# findall() 함수 = 전달받은 결과를 리스트로 반환 
# 리스트로 반환받아야, 결과가 토큰화 됨 = 어절로 이루어짐 = "내가 잘못한거냐" 는 ['내가','잘못한거냐'] 로 저장됨 
# 그리고 리스트 결과를 " ".join()으로 자연스러운 문장으로 합침
dc['글_한글'] = dc['글'].apply(lambda
                           x:" ".join(re.compile("[가-힣]+").findall(x)))

dc.head()
```

## 2. 형태소 분석 

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()

text_s = '나는 회사에 갑니다'
kiwi.tokenize(text_s)
```

```python
result = kiwi.tokenize(text_s)   # text_s 를 토근한 값들 

nouns_s = []                      
for token in result:               # 에 들어있는 단어들을 하나씩 꺼내서
    if token.tag == 'NNG':          # 품사가 '일반명사(NNG)'인 것만 고르고
        nouns_s.append(token.form)   # 그 단어를 리스트에 넣기
nouns_s
```

```python
result = kiwi.tokenize(text_s)  
              
nouns_ss = [token.form for token in result if token.tag=='NNG']
nouns_ss
```

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()

def find_nouns(text): 
    """입력된 문장에서 명사만 추출하여 리스트로 반환"""
    tokens = kiwi.tokenize(text)  

    target_words = []
    for wrd in tokens: 
        if 'NN' in wrd.tag: 
            if len(wrd.form) > 1 : 
                target_words.append(wrd.form)  
    return target_words
    
dc['명사'] = dc['글_한글'].apply(lambda x: find_nouns(x))
dc.head(3)
```

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()

def find_nouns(text): 
    """입력된 문장에서 명사만 추출하여 리스트로 반환"""
    tokens = kiwi.tokenize(text)  
    return [wrd.form for wrd in tokens if 'NN' in wrd.tag and len(wrd.form) > 1]
    
dc['명사'] = dc['글_한글'].apply(lambda x: find_nouns(x))
dc.head(3)
```

```python
# 명사의 형태소로 다 끄집어낸 하나의 리스트  
all_nouns
```

```python
def getGram(tokens, n): 
    """ 분절된 단어를 전달하면 nGram을 반환 """
    bigrams = []
    bigrams = [tokens[i:i + n] for i in range(len(tokens) - n + 1)]       # for문을 컴프리핸션으로 

    ngram_result = pd.DataFrame(data={'nGram': bigrams})
    
    return ngram_result

bigram = getGram(all_nouns, n=2)
bigram
```

```python
def getGram(tokens, n): 
    """ 분절된 단어를 전달하면 nGram을 반환 """
    bigrams = []
    bigrams = [tokens[i:i + n] for i in range(len(tokens) - n + 1)]       # for문을 컴프리핸션으로 

    ngram_result = pd.DataFrame(data={'nGram': bigrams})
    ngram_result['nGram'] = ngram_result['nGram'].apply(lambda x: " ".join(x))
    
    return ngram_result

bigram = getGram(all_nouns, n=2)
bigram
```

```python
def getGram(tokens, n): 
    """ 분절된 단어를 전달하면 nGram을 반환 """
    bigrams = []
    bigrams = [tokens[i:i + n] for i in range(len(tokens) - n + 1)]       # for문을 컴프리핸션으로 

    ngram_result = pd.DataFrame(data={'nGram': bigrams})
    ngram_result['nGram'] = ngram_result['nGram'].apply(lambda x: " ".join(x))
    ngram_result = ngram_result['nGram'].value_counts().to_frame()
    
    return ngram_result

bigram = getGram(all_nouns, n=2)
bigram
```

```python
def getGram(tokens, n): 
    """ 분절된 단어를 전달하면 nGram을 반환 """
    bigrams = []
    bigrams = [tokens[i:i + n] for i in range(len(tokens) - n + 1)]       # for문을 컴프리핸션으로 

    ngram_result = pd.DataFrame(data={'nGram': bigrams})
    ngram_result['nGram'] = ngram_result['nGram'].apply(lambda x: " ".join(x))
    ngram_result = ngram_result['nGram'].value_counts().to_frame()
    ngram_result.reset_index(inplace=True)
    
    return ngram_result

bigram = getGram(all_nouns, n=2)
bigram
```

```python
######## in range(len(tokens) - n + 1) 이해하기 

# tokens이 만약에 ['편의점', '알바', '시간', '음료'] 이고, n = 2 라는 조건문 돌릴 때 

# 결과 
# 편의점 + 알바   --> i = 0 이고 즉 token[0:2]
# 알바 + 시간     --> i = 1 이고 즉 token[1:3]
# 시간 +음료      ---> i = 2 이고 즉 token[2:4]

# 그 뒤에는 없으니  ---> i = 3 과 token[3:5] 는 불가하겠지요. XXXXXXXX 

# 따라서 
# tokens이 만약에 ['편의점', '알바', '시간', '음료'] 이고, n = 2 라는 조건문 돌릴 때 
# len(tokens)가 4
# n = 2 
# ------> len(tokens) - n + 1 
# ------>     4       - 2 + 1 
# ------> 계산해보면     3 이 나옵니다. 
# ------>        range(3) 이란 
# ------>        i가 0, 1, 2 라는 말이겠지요. 
```

```python
necessary_pds = [
    '맥주', '도시락', '음료수', '김밥', '샌드위치', '라면', '치킨', '커피'
]

def label_product(product): 
    """제품 기반으로 문서를 분류"""
    labels = [word for word in necessary_pds if word in product] 
    
    return "-".join(labels)
```

```python
dc['글'] = dc['글'].astype('str')
dc['label'] = dc['글'].apply(lambda x: label_product(x))
dc.head(2)
```

## 주요 키워드 TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

docs = [
    '지금 새벽2시인데 ㅋㅋ 편의점에서 먹을거 추천해주세요! 요새 치킨도 파나?', 
    '편의점 도시락 맛있다! 핫바도 맛있고요!',
    '요새 점심은 그냥 편의점에서 파는 삼각김밥 하나 사서 먹는데 괜찮은듯함. 야식으론 편의점에서 파는 치킨 2조각 사서 먹음ㅋ'
]

tfidf_vect = TfidfVectorizer()

tfidf_vect.fit(docs)
tfidf_array = tfidf_vect.transform(docs).toarray()
tfidf_voca = tfidf_vect.vocabulary_

df_exam = pd.DataFrame(
    data = tfidf_array, 
    columns = tfidf_voca, 
    index = ['문서1', '문서2', '문서3']
)

df_exam
```

```python
dosirak = " ".join(dc[dc['label']=='도시락']['글_한글'].tolist())
noodel = " ".join(dc[dc['label']=='라면']['글_한글'].tolist())
kimbob = " ".join(dc[dc['label']=='김밥']['글_한글'].tolist())
coffee = " ".join(dc[dc['label']=='커피']['글_한글'].tolist())

docs = [dosirak, noodel,kimbob, coffee]
```

```python
tfidf = TfidfVectorizer().fit(docs)
tfidf_array = tfidf.transform(docs).toarray()
tfidf_voca = tfidf.vocabulary_

df_tfidf = pd.DataFrame(
    tfidf_array,
    columns = tfidf_voca,
    index=['도시락', '라면', '김밥', '커피']
)

df_tfidf.T
```
